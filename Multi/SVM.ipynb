```python
from sklearn.feature_selection import chi2
import pandas as pd
import numpy as np

# Assuming df1, df2, df3 are your DataFrames with "SampleID" as the common ID column

# Step 1: Perform feature selection separately for each dataset
def perform_feature_selection(df, num_features):
    X = df.iloc[:, 2:]  # Assuming features start from the 3rd column
    y = df["histological_type"]

    # Calculate chi-squared scores for each feature
    chi2_scores, _ = chi2(X, y)

    # Select the top k features based on chi-squared scores
    selected_features_idx = np.argsort(chi2_scores)[-num_features:]
    selected_features = X.columns[selected_features_idx]

    return df[['SampleID', 'histological_type'] + list(selected_features)].copy(), selected_features

# Select the desired number of features for each dataset
num_features_df1 = 70
num_features_df2 = 60
num_features_df3 = 60

df1_selected, features_df1 = perform_feature_selection(df1, num_features_df1)
df2_selected, features_df2 = perform_feature_selection(df2, num_features_df2)
df3_selected, features_df3 = perform_feature_selection(df3, num_features_df3)

# Step 2: Combine the selected features based on the common "SampleID" column
combined_df = pd.merge(df1_selected, df2_selected, on='SampleID', how='inner')
combined_df = pd.merge(combined_df, df3_selected, on='SampleID', how='inner')

# Step 3: Order the combined features based on their Chi-squared values
combined_chi2_scores, _ = chi2(combined_df.iloc[:, 2:], combined_df["histological_type"])
ordered_features_idx = np.argsort(combined_chi2_scores)[::-1]  # Reverse order for descending

# Step 4: Select the top 90 features and create a new dataset
num_selected_features = 91
selected_features = combined_df.columns[ordered_features_idx][:num_selected_features]
final_dataset = combined_df[['SampleID', 'histological_type'] + list(selected_features)].copy()

# Count the occurrences of features from each dataset in the final 90
count_df1 = sum(feature in features_df1 for feature in final_dataset.columns)
count_df2 = sum(feature in features_df2 for feature in final_dataset.columns)
count_df3 = sum(feature in features_df3 for feature in final_dataset.columns)

# Print the count of selected features for each dataset
print(f"Count of features from df1 in the final dataset: {count_df1}")
print(f"Count of features from df2 in the final dataset: {count_df2}")
print(f"Count of features from df3 in the final dataset: {count_df3}")

# Display the final dataset
print(final_dataset.columns)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.feature_selection import chi2
from sklearn.model_selection import cross_val_score
from sklearn.utils import resample
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming df is your DataFrame

svm_classifier = SVC(kernel='linear', random_state=42, C=1)  # Use linear kernel for simplicity
K = np.arange(start=5, stop=90, step=5)
iterations = 10  # Number of iterations
feature_numbers = []
mean_accuracies = []
std_accuracies = []

# Loop through the selected features
for idx, k in enumerate(K):
    accuracies_iter = []

    for iteration in range(iterations):
        # Feature selection using Chi-squared
        X = df.iloc[:, 2:92]
        y = df["histological_type"]

        # Calculate chi-squared scores for each feature
        chi2_scores, _ = chi2(X, y)

        # Select the top k features based on chi-squared scores
        selected_features_idx = np.argsort(chi2_scores)[-k:]
        selected_features = X.columns[selected_features_idx]

        df_selected = df[['SampleID', 'histological_type'] + list(selected_features)].copy()

        # Undersample the dataset with a random seed between 0 and 100 for each iteration
        undersample_seed = np.random.randint(101)  # Generate a random seed between 0 and 100
        cervical_squamous_df = df_selected[df_selected['histological_type'] == 0]
        if len(cervical_squamous_df) > 47:
            cervical_squamous_df = resample(cervical_squamous_df, n_samples=47, random_state=undersample_seed)
        adeno_df = df_selected[df_selected['histological_type'] == 1]
        df2 = pd.concat([cervical_squamous_df, adeno_df])

        # Split the data into features (X) and target (y)
        X_selected = df2.iloc[:, 2:]
        y_selected = df2["histological_type"]

        # Perform 5-fold cross-validation
        scores = cross_val_score(svm_classifier, X_selected, y_selected, cv=5, scoring='accuracy')

        # Calculate average accuracy and standard deviation across folds
        accuracy = np.mean(scores)
        std_accuracy = np.std(scores)
        accuracies_iter.append(accuracy)

        # Print update after each iteration
        print(f"Iteration {iteration + 1}/{iterations} for k={k}: Accuracy = {accuracy:.4f}, Std Dev = {std_accuracy:.4f}, Undersample Seed = {undersample_seed}")

    # Store the feature number and its mean accuracy and standard deviation across iterations
    feature_numbers.append(k)
    mean_accuracies.append(np.mean(accuracies_iter))
    std_accuracies.append(np.std(accuracies_iter))

    # Print update after each feature number
    print(f"Finished iteration for k={k}: Mean Accuracy = {mean_accuracies[-1]:.4f}, Mean Std Dev = {std_accuracies[-1]:.4f}\n")

# Plotting the feature numbers, mean accuracies, and standard deviations
plt.figure(figsize=(10, 6))
plt.errorbar(feature_numbers, mean_accuracies, yerr=std_accuracies, fmt='o-', label='Accuracy', capsize=5, markersize=8)

# Annotate each point with its accuracy and std inside a colored box
for i in range(len(feature_numbers)):
    box_properties = dict(boxstyle='round', facecolor='lightgray', alpha=0.7)
    plt.text(
        feature_numbers[i],
        mean_accuracies[i],
        f'{mean_accuracies[i]:.4f}\nÂ±{std_accuracies[i]:.4f}',
        fontsize=8,
        ha='right',
        va='bottom',
        bbox=box_properties
    )

plt.title('Mean Accuracy with Std Dev for each Feature Number (Chi-squared + Undersampling + SVM with 5-Fold Cross-Validation)')
plt.xlabel('Number of Features')
plt.ylabel('Mean Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
