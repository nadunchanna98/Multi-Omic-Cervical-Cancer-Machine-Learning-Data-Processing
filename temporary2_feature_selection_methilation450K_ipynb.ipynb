## FOR chi-square

```python
from scipy.stats import chi2_contingency

df_new7 = df_new6.copy()

# Calculate the chi-square test statistics and p-values for each feature
chi2_stats = []
p_values = []
for feature in df_new7.columns:
    contingency_table = pd.crosstab(df_new7[feature], df_new7['histological_type'])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    chi2_stats.append(chi2)
    p_values.append(p_value)

# # Create a DataFrame with the chi-square statistics and p-values
# chi2_df = pd.DataFrame({'Feature': df_new7.columns, 'Chi2_Stat': chi2_stats, 'P-Value': p_values})

# # Sort the DataFrame based on p-values in ascending order
# sorted_chi2_df = chi2_df.sort_values(by='P-Value', ascending=False)

# # Select the top 200 features based on p-values
# top_200_features = sorted_chi2_df['Feature'][:200]

# # Extract the selected features from the DataFrame
# selected_features_df = df[top_200_features]
# selected_features_df
```

```python
# Sort the features based on p-values
sorted_indices = sorted(range(len(p_values)), key=lambda i: p_values[i])
sorted_p_values = [p_values[i] for i in sorted_indices]
sorted_features = [df_new7.columns[i] for i in sorted_indices]

# Select the first 200 features and their p-values
top_200_p_values = sorted_p_values[:200]
top_200_features = sorted_features[:200]

# Create a plot
plt.figure(figsize=(12, 6))
plt.barh(range(len(top_200_p_values)), top_200_p_values, tick_label=top_200_features)
plt.xlabel('P-Value')
plt.title('P-Values for the Top 200 Features')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest p-values at the top
plt.show()
```

```python
# Sort the features based on chi-square values
sorted_indices = sorted(range(len(chi2_stats)), key=lambda i: chi2_stats[i])
sorted_chi2_values = [chi2_stats[i] for i in sorted_indices]
sorted_features = [df_new7.columns[i] for i in sorted_indices]

# Select the features in the range 40 to 60 and their chi-square values
start_index = 40
end_index = 50
selected_chi2_values = sorted_chi2_values[-end_index:-start_index][::-1]  # Reverse to get highest values first
selected_features = sorted_features[-end_index:-start_index][::-1]  # Reverse to match the order of chi2 values

# Create a plot
plt.figure(figsize=(10, 6))
plt.barh(range(len(selected_chi2_values)), selected_chi2_values, tick_label=selected_features)
plt.xlabel('Chi-Square Value')
plt.title('Chi-Square Values for Features 40 to 60')
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest chi-square values at the top

# Adjust y-axis label position and rotation for readability
plt.tick_params(axis='y', labelrotation=0)
plt.tight_layout()

plt.show()
```
