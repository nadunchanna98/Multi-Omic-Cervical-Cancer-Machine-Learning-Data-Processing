```python
df1 = df1.drop(['Unnamed: 0'], axis=1)
```

```python
#change id name
df1 = df1.rename(columns={'id': 'SampleID'})
df1.head()
```

```python
from sklearn.feature_selection import chi2
import pandas as pd
import numpy as np

# Assuming df1, df2, df3 are your DataFrames with "SampleID" as the common ID column

# Step 1: Perform feature selection separately for each dataset
def perform_feature_selection(df, num_features):
    X = df.iloc[:, 2:]  # Assuming features start from the 3rd column
    y = df["histological_type"]

    # Calculate chi-squared scores for each feature
    chi2_scores, _ = chi2(X, y)

    # Select the top k features based on chi-squared scores
    selected_features_idx = np.argsort(chi2_scores)[-num_features:]
    selected_features = X.columns[selected_features_idx]

    # Create a DataFrame with feature names and corresponding chi-square values
    feature_selection_df = pd.DataFrame({
        'Feature_name': selected_features,
        'Chi_Square_Value': chi2_scores[selected_features_idx]
    })

    return feature_selection_df


# Select the desired number of features for each dataset
num_features_df1 = 70
num_features_df2 = 60
num_features_df3 = 60

# Create new DataFrames for each dataset
methilation450K = perform_feature_selection(df1, num_features_df1)
HiSeqv = perform_feature_selection(df2, num_features_df2)
miRNA = perform_feature_selection(df3, num_features_df3)

# Rename columns in each DataFrame
methilation450K.columns = ['methilation450K_Feature_name', 'methilation450K_Chi_Square_Value']
HiSeqv.columns = ['HiSeqv_Feature_name', 'HiSeqv_Chi_Square_Value']
miRNA.columns = ['miRNA_Feature_name', 'miRNA_Chi_Square_Value']

# Combine DataFrames into a single DataFrame
result_df = pd.concat([methilation450K, HiSeqv, miRNA], axis=1)

# Display the combined DataFrame
result_df
```
