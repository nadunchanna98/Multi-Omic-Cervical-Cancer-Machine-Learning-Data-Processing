```python
column_names=clinical.columns.tolist()

for column in column_names:
    print(column)
```

```python
columns_to_extract = ['sampleID','histological_type']

new_clinical = clinical[columns_to_extract].copy()
new_clinical
```

```python
unique_values = new_clinical['histological_type'].unique()
unique_values
```

```python
column_values = new_clinical['histological_type']

# Calculate the total count of values in the column
total_count = column_values.count()

# Calculate the value counts for each unique value in the column
value_counts = column_values.value_counts()

# Calculate the percentage for each value
percentages = (value_counts / total_count) * 100

# Display the percentages
print(percentages)
```

```python
transposed_df  = df.transpose()
transposed_df.columns=transposed_df.iloc[0]
transposed_df
```

```python
transposed_df = transposed_df.drop("sample", axis=0)
```

```python
transposed_df
```

```python
print(transposed_df.columns)
```

```python
transposed_df.reset_index(inplace=True)
transposed_df
```

```python
merged_df = pd.merge(transposed_df,new_clinical, left_on='index', right_on='sampleID')
```

```python
print(merged_df.columns)
```

```python
print(merged_df[['index', 'sampleID','histological_type']])
```

```python
merged_df
```

```python
# Change the name of a column in the DataFrame
merged_df1 = merged_df.rename(columns={'index': 'SampleID'})
merged_df1
```

```python
#To swap the position of the 'histological_type' column and move it as the second column
columns = merged_df1.columns.tolist()

# Find the index of 'histological_type' column
histological_type_index = columns.index('histological_type')

# Move the 'histological_type' column to the second position
columns = [columns[0]] + [columns[histological_type_index]] + columns[1:histological_type_index] + columns[histological_type_index+1:]

# Reorder the DataFrame columns
merged_df1 = merged_df1[columns]
merged_df1
```

```python
print(merged_df1.columns)
```

```python
data = merged_df1.drop("sampleID", axis=1)
```
